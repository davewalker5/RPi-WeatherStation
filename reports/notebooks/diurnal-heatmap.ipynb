```python
# Load the readings for each sensor and produce a combined data frame
bme280_df = load_sensor_readings("bme280")
veml7700_df = load_sensor_readings("veml7700")
sgp40_df = load_sensor_readings("sgp40")
combined_df = merge_sensor_readings([bme280_df, veml7700_df, sgp40_df])
combined_df.head()
```

```python
# Add hour-of-day column (0â€“23) and a date column
combined_df["hour"] = combined_df.index.hour
combined_df["date"] = combined_df.index.date
```

```python
# Get the export folder path
export_folder_path = get_export_folder_path()
```

# Diurnal Temperature Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns

pivot = combined_df.pivot_table(
    index="hour",
    columns="date",
    values="temperature",
    aggfunc="mean"
)

plt.figure(figsize=(14,6))
sns.heatmap(pivot, cmap="YlOrRd")
plt.title("Temperature Heatmap (Hour vs Day)")
plt.xlabel("Date")
plt.ylabel("Hour of Day")

# Export to PNG or PDF, if required
export_chart(export_folder_path, "diurnal_temperature_heatmap", "png")

plt.show()
```
